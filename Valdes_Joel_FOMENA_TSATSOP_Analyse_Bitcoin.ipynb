**Chargement des donnees**

```python
# Import des librairies pandas et numpy.
import pandas as pd
import numpy as np
```

```python
# Lecture des données stockés au format csv

btc = pd.read_csv(
    "BTC_Historical_Data.csv", sep=';', decimal=','
)
```

```python
#J'effectue un filtre pour la variable close pour eleminer les donnees de la variable egal a 0
btc = btc[(btc[['Close']] != 0).all(axis=1)]
```

```python
# Formatage du champ date
btc["Date"] = pd.to_datetime(
      btc["Date"]
    #, format="%Y-%m-%d", errors="coerce"
)
# Extraire uniquement la date sans l'heure
#btc['Date'] = btc['Date'].dt.date


# Afficher les premières lignes du DataFrame pour vérifier


btc
```

```python
#definir la date comme index
btc.set_index('Date', inplace=True)
```

```python
#affichons les colonnes close et volume
btc_subset = btc[['Close', 'Volume']]
btc_subset.head(15)
```

**correlation close et volume**

```python
# Calcul de la corrélation entre "Close" et "Volume"

btc_monthly = btc_subset.resample('ME').mean()
correlation = btc_subset['Close'].corr(btc['Volume'])

print(f"Corrélation entre Close et Volume : {correlation}")
```

La corrélation entre les variables "Close" et "Volume" est de 0.726, ce qui indique une relation positive modérément forte entre ces deux variables

**Representation graphique de close et volume avec bokeh**


```python
#importons la bibliotheque bokeh
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource
from bokeh.io import output_notebook

# Affichage des graphiques directement dans le notebook
output_notebook()
```

```python
from bokeh.models import ColumnDataSource, RangeTool
from bokeh.plotting import figure, show
from bokeh.layouts import column
import pandas as pd

# Vérifier que l'index est bien de type datetime
if not isinstance(btc.index, pd.DatetimeIndex):
    btc.index = pd.to_datetime(btc.index)

# Préparer la source de données
source = ColumnDataSource(data=dict(date=btc.index, close=btc['Close'], volume=btc['Volume']))

# Création du graphique pour Volume
p2 = figure(height=300, width=800, tools="xpan,wheel_zoom,box_zoom,reset",
            toolbar_location="above", x_axis_type="datetime",
            x_axis_location="above", background_fill_color="#efefef",
            x_range=(btc.index[0], btc.index[-1]))

p2.line('date', 'volume', source=source, line_width=2, color="green", legend_label="Volume")
p2.yaxis.axis_label = 'Volume'

# Ajout de la boîte de sélection
select = figure(title="Sélectionner la plage", height=130, width=800,
                y_range=p2.y_range, x_axis_type="datetime",
                y_axis_type=None, tools="", toolbar_location=None,
                background_fill_color="#efefef")

range_tool = RangeTool(x_range=p2.x_range)
range_tool.overlay.fill_color = "navy"
range_tool.overlay.fill_alpha = 0.2

select.line('date', 'volume', source=source, line_width=2, color="green")
select.ygrid.grid_line_color = None
select.add_tools(range_tool)

# Affichage des graphiques en colonne
show(column(p2, select))
```

```python
from bokeh.models import ColumnDataSource, RangeTool
from bokeh.plotting import figure, show
from bokeh.layouts import column
import pandas as pd

# Vérifier que la colonne 'Date' est bien l'index et de type datetime
if not isinstance(btc.index, pd.DatetimeIndex):
    btc.index = pd.to_datetime(btc.index)

# Préparer la source de données
source = ColumnDataSource(data=dict(date=btc.index, close=btc['Close'], volume=btc['Volume']))

# Création du premier graphique pour Close avec zoom
p1 = figure(height=300, width=800, tools="xpan,wheel_zoom,box_zoom,reset",
            toolbar_location="above", x_axis_type="datetime",
            x_axis_location="above", background_fill_color="#efefef",
            x_range=(btc.index[0], btc.index[-1]))

p1.line('date', 'close', source=source, line_width=2, color="blue", legend_label="Close")
p1.yaxis.axis_label = 'Close'

# Ajout de la boîte de sélection
select = figure(title="Sélectionner la plage", height=130, width=800,
                y_range=p1.y_range, x_axis_type="datetime",
                y_axis_type=None, tools="", toolbar_location=None,
                background_fill_color="#efefef")

range_tool = RangeTool(x_range=p1.x_range)
range_tool.overlay.fill_color = "navy"
range_tool.overlay.fill_alpha = 0.2

select.line('date', 'close', source=source, line_width=2, color="blue")
select.ygrid.grid_line_color = None
select.add_tools(range_tool)

# Affichage des graphiques en colonne
show(column(p1, select))
```

```python
from bokeh.models import HoverTool
# Créer la source de données pour Bokeh
source = ColumnDataSource(btc)

# Créer la figure
p = figure(title="Corrélation entre Close et Volume au fil du temps",
           x_axis_label="Date",
           y_axis_label="Volume",
           width=800, height=500,
           x_axis_type='datetime')  # Définir l'axe X comme étant de type datetime

# Ajouter le scatter plot avec la Date sur l'axe x et Volume sur l'axe y
p.scatter(x='Date', y='Volume', source=source, size=7, color="navy", alpha=0.6)

# Ajouter les tooltips pour afficher la Date, Close et Volume
tooltips = [
    ("Date", "@Date{%F}"),  # Afficher la date au format yyyy-mm-dd
    ("Close", "@Close{0.2f}"),
    ("Open", "@Open{0.2f}"),
    ("Volume", "@Volume{0.2f}")
]

# Ajouter le HoverTool pour afficher les tooltips
hover = HoverTool(tooltips=tooltips, formatters={'@Date': 'datetime'})
p.add_tools(hover)

# Affichage
output_notebook()  # Afficher le graphique directement dans un notebook Jupyter
show(p)
```

**Creation d'une serie mensuelle et annuelle sur la moyenne**

```python
#serie mensuelle
btc_monthly = btc_subset.resample('ME').mean().interpolate(method='linear')

#serie annuelle
btc_annual = btc_subset.resample('YE').mean().interpolate(method='linear')

btc_monthly.head()
btc_annual.head()
```

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource, RangeTool
from bokeh.layouts import column

# Création de la série temporelle pour 'Close' et 'Volume'
dates = np.array(btc_monthly.index, dtype=np.datetime64)
source_close = ColumnDataSource(data=dict(date=dates, variable=btc_monthly['Close']))
source_volume = ColumnDataSource(data=dict(date=dates, variable=btc_monthly['Volume']))

# Graphique pour 'Close'
p_close = figure(height=300, width=800, tools="xpan", toolbar_location=None,
                 x_axis_type="datetime", x_axis_location="above",
                 background_fill_color="#efefef", x_range=(btc_subset.index[1500], btc_subset.index[2500]))

p_close.line('date', 'variable', source=source_close, color="blue")
p_close.yaxis.axis_label = 'Close'


# Création de la boîte de sélection
select = figure(title="Faites glisser le milieu et les bords de la boîte de sélection pour modifier la plage ci-dessus.",
                height=130, width=800, y_range=p_close.y_range,
                x_axis_type="datetime", y_axis_type=None,
                tools="", toolbar_location=None, background_fill_color="#efefef")

range_tool = RangeTool(x_range=p_close.x_range)
range_tool.overlay.fill_color = "navy"
range_tool.overlay.fill_alpha = 0.2

select.line('date', 'variable', source=source_close, color="blue")
select.ygrid.grid_line_color = None
select.add_tools(range_tool)


# Affichage du graphique
show(column(p_close, select))
```
